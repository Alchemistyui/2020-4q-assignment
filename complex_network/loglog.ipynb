```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.utils.random_sequence import powerlaw_sequence
from IPython import display
%matplotlib inline
display.set_matplotlib_formats('svg')

er = nx.erdos_renyi_graph(10000, 0.001)
print("Erdos-Renyi graph")
print(nx.info(er))
print('Number of connected components: ', nx.number_connected_components(er))
print('Number of triangles: ', sum(nx.triangles(er).values()) / 3)
print('Transitivity: ', nx.transitivity(er))
print("Maximum degree:", max(dict(er.degree()).values()))
print("Minimum degree:", min(dict(er.degree()).values()))
plt.subplot(221)
plt.plot(nx.degree_histogram(er))
print()
ba = nx.barabasi_albert_graph(10000, 5)
print("Barabasi-Albert graph")
print(nx.info(ba))
print('Number of connected components: ', nx.number_connected_components(ba))
print('Number of triangles: ', sum(nx.triangles(ba).values()) / 3)
print('Transitivity: ', nx.transitivity(ba))
print("Maximum degree:", max(dict(ba.degree()).values()))
print("Minimum degree:", min(dict(ba.degree()).values()))
plt.subplot(222)
plt.plot(nx.degree_histogram(ba))

# plot logarithmic axes
plt.subplot(223)
plt.xscale("log")
plt.yscale("log")
plt.grid(which="both")
plt.plot(nx.degree_histogram(er))
plt.subplot(224)
plt.xscale("log")
plt.yscale("log")
plt.grid(which="both")
plt.plot(nx.degree_histogram(ba))
```
