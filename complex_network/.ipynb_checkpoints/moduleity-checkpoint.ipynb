```python
import networkx as nx
from matplotlib import pyplot as plt
from IPython import display
%matplotlib inline
display.set_matplotlib_formats('svg')

# https://python-louvain.readthedocs.io/en/latest/
!pip install python-louvain # for installing python-louvain
from community import community_louvain
import community

G = nx.karate_club_graph()
color_map = ['red'] * nx.number_of_nodes(G)
membership = [[], []]
dict = {}

print("==============================")
print("real partition")
for i in range(nx.number_of_nodes(G)) :
  if G.nodes[i]['club'] == 'Officer' :
    color_map[i] = 'yellow'
    membership[0].append(i)
    dict[i] = 0
  else :
    membership[1].append(i)
    dict[i] = 1
print("membership:", membership)
print(dict)
print("modularity:", community.modularity(dict, G))
print(color_map)
plt.subplot(121)
nx.draw_spring(G, node_size=400, node_color=color_map, with_labels=True, font_weight='bold')

print("==============================")
print("partition based on modularity")
partition = community_louvain.best_partition(G)

#membership_b  = [[]] * (max(partition.values()) + 1) NG (all objects in the list will be the same)
m = max(partition.values()) + 1
membership_b  = [[] * m for i in range(m)]

for k, v in partition.items() :
  membership_b[v].append(k)
print("membership:", membership_b)
print(partition)
print("modularity:", community.modularity(partition, G))

colors = ['red', 'blue', 'green', 'yellow', 'black']
color_map_b = ['red'] * nx.number_of_nodes(G)
for i in range(nx.number_of_nodes(G)) :
  color_map_b[i] = colors[partition[i]]
print(color_map_b)
plt.subplot(122)
nx.draw_spring(G, node_size=400, node_color=color_map_b, with_labels=True, font_weight='bold')
```

## Quiz 6-2 Inputs and Outputs of Modularity Q
$$
Q=\frac{1}{2 m} \sum_{i j}\left(A_{i j}-\frac{k_{i} k_{j}}{2 m}\right) \delta\left(c_{i}, c_{j}\right)
$$

where $A_{i j}-\frac{k_{i} k_{j}}{2 m}$ means the difference between the actual number of edges between nodes and the expected number of edges between them. And $A_{i j}$ represents the weight of the edge between i and j, $k_i=\sum_j A_{ij}$ is the sum of
the weights of the edges attached to vertex i, $c_i$ is the community to which vertex i is assigned, $\delta(u, v)$ is 1 if $u=v$ and 0 otherwise and $m=\frac{1}{2} \sum_{i j} A_{i j}$.

So for this formula, we need inputs: $A_{ij}$ and $c_i$, in other words, we need the original graph and its partition. And the output is Modularity.

## What the outputs mean

Modularity is one measure of the structure of networks, which is widely used as a measure for how good
a clustering is.
Networks with high modularity have dense connections between the nodes within modules but sparse connections between nodes in different modules.

According to Newman(2008): "Modularity is strictly less than 1, and takes positive values if there are more edges between vertices of the same type than we would expect by chance, and negative ones if there are less".
