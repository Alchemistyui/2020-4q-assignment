```python

import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

G = nx.karate_club_graph()
plt.figure(figsize=(6, 6))
nx.draw_spring(G, node_size=400, node_color='red', with_labels=True, font_weight='bold')

print("n =", nx.number_of_nodes(G))
print("m =", nx.number_of_edges(G))
A = nx.adjacency_matrix(G).todense()
print(A)
print(A*A)

print("sum", np.sum(A))
print("trace", np.trace(A))
```

# Exercise 1: Answer is 45

## Sort the vertices according to their degree(from largest to smallest)

```python
# degree for all the vertices
degree = A.sum(0)
# get the index(vertices serial number) from largest to smallest by sorting
sort_idx = np.argsort(A.sum(0)).A[0]
sort_idx
```

## Make a neighborhood set N(v) for each vertices v.
make sure that vertices u in N(v) is connected with v, and the degree of u is larger than v's.

By this way, we can reduce the calculated amount and make sure we don't count triangles repeatedly

```python
# whole neighbor set
N = {}

for v in range(nx.number_of_nodes(G)):
    # get all the connected vertices
    connect = np.where(A[v] == 1)[1]
    # get the vertices whose degree larger than v
    larger = sort_idx[np.where(sort_idx == v)[0].item()+1:]
    n_v = np.intersect1d(connect,larger)
    N[v] = n_v
N
```

## Traverse every edges

For each edge (u,v), check if there are intersection vertices in the neighborhood set in u and v.

Add up all the intersections for every edges and then we can get the number of triangles.

```python
# get the Upper triangle of A to prevent double counting edges
U_A = np.triu(A,1)
# get all the edges
edges = np.argwhere(U_A)
```

```python
# add up all the intersections for every edges
sum = 0
for e in edges:
    v1, v2 = e
    inter = np.intersect1d(N[v1],N[v2])
    sum += len(inter)
print('Number of triangles: {}'.format(sum))
```

# Exercise 2: Answer is 84

In order to get the maximum number of triangles, we 
In order to get the maximum number, any two of all 9 vertices should not be on the same line. Therefore, select any three points can form a triangle, so the number of triangles is: $C_9^3 = 84$.

```python
import itertools

G = nx.Graph()
# initial all the vertices
vertices = [i for i in range(9)]
G.add_nodes_from(vertices)
# all the triangles generated by the combination
triangles = list(itertools.combinations(range(9),3))
print('Maximum number of triangles: {}'.format(len(triangles)))
# get all the edges
edges = []
for tri in triangles:
    e = list(itertools.combinations(tri,2))
    edges.extend(e)
G.add_edges_from(edges)
plt.figure(figsize=(6, 6))
nx.draw_spring(G, node_size=400, node_color='pink', with_labels=True)
```
